```python
import cv2
import os

# Créer un dossier pour stocker les images
folder_name = "expressions"
if not os.path.exists(folder_name):
    os.makedirs(folder_name)

# Définir les paramètres de la caméra
camera = cv2.VideoCapture(0)
camera.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
camera.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

# Définir les expressions à capturer
expressions = ["main", "point"]

# Capturer 15 images pour chaque expression
for expression in expressions:
    print(f"Capture de l&#039;expression {expression}...")
    expression_folder = os.path.join(folder_name, expression)
    if not os.path.exists(expression_folder):
        os.makedirs(expression_folder)
    for i in range(100):
        ret, frame = camera.read()
        image_name = f"{expression}_{i}.jpg"
        image_path = os.path.join(expression_folder, image_name)
        cv2.imwrite(image_path, frame)
        cv2.imshow("Image", frame)
        cv2.waitKey(100)
    print(f"Capture de l&#039;expression {expression} terminée.")

# Fermer la caméra et détruire les fenêtres
camera.release()
cv2.destroyAllWindows()
```
