```python
import pandas as pd
import numpy as np
```

```python
url="C:/Users/user/Desktop/Data.xlsx"
df=pd.read_excel(url)
```

# logistique regression

```python
df
```

```python
df.isna().sum()
```

```python
df=df.dropna()
```

```python
df.isna().sum()
```

```python
df=df.drop(columns='Date')
df=df.drop(columns='Visibilityclass')
```

```python
df['Age'] = df['Age'].astype(int)
df['Course'] = df['Course'].astype(int)
```

```python
df['Day'] = df['Day'].astype('category')
df['Day'] = df['Day'].cat.codes

df['Vehicle'] = df['Vehicle'].astype('category')
df['Vehicle'] = df['Vehicle'].cat.codes

df['Cause'] = df['Cause'].astype('category')
df['Cause'] = df['Cause'].cat.codes

df['Cause_Cat'] = df['Cause_Cat'].astype('category')
df['Cause_Cat'] = df['Cause_Cat'].cat.codes
```

```python
df
```

```python
X=df.drop(columns='Accident')
X
```

```python
y=df["Accident"]
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=21)
```

```python
x_train.sample(5)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)
```

```python
x_test_scaled
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(random_state=0,
                            C=1,
                            fit_intercept=True).fit(x_train_scaled,y_train)
```

```python
log_reg.predict(x_test_scaled)  #compare avec np.array(y_test)
```

```python
np.array(y_test)
```

```python
log_reg.score(x_test_scaled,y_test)  #score is 0.9883720930232558
```
